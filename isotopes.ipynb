```python
import numpy as np
from numpy.fft import fft, ifft
from matplotlib import pyplot as plt
```

```python
a = np.array([1,6,3,1,0.5])
b = np.convolve(a,a)
for i in range(1):
    b = np.convolve(a,b)
c = fft(np.concatenate((np.zeros(a.size),a,np.zeros(a.size))))
bb = ifft(c**3)
type_data = 'int'
cc = bb.astype(type_data)
dd = np.roll(bb.astype(type_data), a.size)
print(b.astype(type_data))
print(cc) # 
print(dd) # 
plt.plot(b)
plt.plot(cc)
#plt.plot(bb)
plt.show()
```
